```python
from bs4 import BeautifulSoup
import requests, lxml, os, json
from dateutil.parser import parse
import pandas as pd

proxies = {
    'http': os.getenv('HTTP_PROXY')  # or just type proxy here without os.getenv()
}

headers = {
    'User-agent':
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36 Edge/18.19582"
}

df = pd.DataFrame()

counter = 0

for i in range(0, 112):
    start = str(i * 10)

    if i == 0:
        url = 'https://citeseerx.ist.psu.edu/search?q=smart+health&submit.x=0&submit.y=0&sort=rlv&t=doc'
    else:
        url = 'https://citeseerx.ist.psu.edu/search?q=smart+health&t=doc&sort=rlv&start=' + start

    html = requests.get(url, headers=headers, proxies=proxies).text
    soup = BeautifulSoup(html, 'lxml')

    for i in soup.findAll("div", {'class': "result"}):
        abstract = i.find("div", {'class': "pubabstract"}).text

        d = {
            '500 abstracts': abstract
        }

        df = df.append(d, ignore_index=True)


df.to_csv("citeseer")
print("500 records download successful to Google Colab folders")
```

```python
import nltk
nltk.download()
```

```python
df['Lower'] = df['500 abstracts'].apply(lambda i: " ".join(i.lower() for i in i.split()))
```

```python
from textblob import TextBlob
df['Spellings'] = df['Lower'].apply(lambda i: str(TextBlob(i).correct()))
```

```python
import nltk
nltk.download('punkt')
df['tok'] = df['Spellings'].apply(lambda i: TextBlob(i).words)
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
df['post stem'] = df['tok'].apply(lambda i: " ".join([stemmer.stem(word) for word in i]))
```

```python
from textblob import Word
import nltk
nltk.download('wordnet')

df['post lem'] = df['post stem'].apply(lambda i: " ".join([Word(word).lemmatize() for word in i.split()]))
df.to_csv('/content/sample_data/500abstracts.csv',index=False)

df.to_csv('file1.csv')

```

```python
import itertools
import scipy
import numpy
import collections
from nltk import ngrams
from nltk.tokenize import word_tokenize

lines = []
for line in df['post lem']:
  lines.append(word_tokenize(line))
postclean = [i for i in lines if i != []]
values = list(itertools.chain.from_iterable(postclean))
```

```python
trigrams = nltk.trigrams(values)
fdist = nltk.FreqDist(trigrams)
fdist
```

```python
from collections import Counter
bigrams = nltk.bigrams(values)
fdist = nltk.FreqDist(bigrams)
bigrams_dict = dict(fdist)
for word in bigrams_dict:
  print( str(word) + ':' + str(bigrams_dict[word] / values.count(word[0])))
```

# **Question 2: Undersand TF-IDF and Document representation**

(40 points). Starting from the documents (all the reviews, or abstracts, or tweets) collected for assignment two, write a python program: 

(1) To build the **documents-terms weights (tf*idf) matrix bold text**.

(2) To rank the documents with respect to query (design a query by yourself, for example, "An Outstanding movie with a haunting performance and best character development") by using **cosine similarity**.

```python
import math
def tf_idf(line,dot):
  s = len(line.split(" "))
  tf = line.count(dot)/s
  idf = 0
  if(tf!=0):
    idf = math.log(s)/line.count(dot)
  else:
    return 0;
  return tf*idf

lines = df["Spellings"].values.tolist()
phrases_list = set([j for i in lines for j in i.split(" ")])
tfidf = pd.DataFrame(phrases_list,columns=["phrases"])
c = 0
for i in lines:
  tfidf[str(c)] = tfidf["phrases"].apply(lambda x : tf_idf(i,x))
  c+=1
print(tfidf)
```

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
```

```python
query = 'Smart cards are used in information technologies as portable integrated devices with data storage and data processing capabilities. As in other fields, smart card use in health systems became popular due to their increased capacity and performance. Smart cards are used as a Electronic'
list1 = word_tokenize(query)
docs = stopwords.words('english') 
set1 = {y for y in list1 if not y in docs}
```

```python
def c_v(set1, set2):
  r_v = set1.union(set2)
  lista =[]; listb =[]
  for y in r_v: 
      if y in set1: lista.append(1)
      else: lista.append(0) 
      if y in set2: listb.append(1) 
      else: listb.append(0)
  return r_v, lista, listb
```

```python
def c_cosine(r_v, lista, listb):
  count = 0
  for b in range(len(r_v)): 
        count+= lista[b]*listb[b] 
  cosine = count / float((sum(lista)*sum(listb))**0.5) 
  return cosine
```

```python
samelist = []
for line in df['Spellings']:
  list2 = word_tokenize(line)
  set2 = {y for y in list2 if not y in docs}
  r_v, lista, listb = c_v(set1, set2)
  try:
    similarity = c_cosine(r_v, lista, listb)
  except ZeroDivisionError:
    similarity = 'None'
  samelist.append(similarity)
cosine_df = pd.DataFrame(list(zip(df['500 abstracts'],samelist)), columns=['500 abstracts','Cosine Similarity'])
cosine_df
```
